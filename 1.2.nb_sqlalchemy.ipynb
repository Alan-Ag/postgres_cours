# SQLAlchemy

```python
import pandas as pd
from sqlalchemy import create_engine, text

# Database connection details
DB_NAME = 'commerce_3'
DB_USER = 'postgres'
DB_PASSWORD = '025858'
DB_HOST = 'localhost'
DB_PORT = '5432'

# Create SQLAlchemy engine
engine = create_engine(f'postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}')
```

```python
# Read data from database
sql = text('SELECT * FROM clients')
clients = pd.read_sql(sql, engine)
clients
```

```python
sql = text('SELECT * FROM produits')
produits = pd.read_sql(sql, engine)
produits
```

```python
sql = text('SELECT * FROM commandes')
commandes = pd.read_sql(sql, engine)
commandes
```

# Insérer un nouveau client

```python
query = text("INSERT INTO clients (nom, prenom) VALUES ('Zidane', 'Zinedine')")
# execute the query
with engine.connect() as connection:
    connection.execute(query)
    connection.commit() 
```

```python
sql = text('SELECT * FROM clients')
clients = pd.read_sql(sql, engine)
clients
```

```python
nom = 'MBappe'
prenom = 'Kylian'

query = text("INSERT INTO clients (nom, prenom) VALUES (:nom, :prenom)")
params = {'nom': nom, 'prenom': prenom}
with engine.connect() as connection:
    connection.execute(query, params)
    connection.commit()
```
